```python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import matplotlib.patches as mpatches
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({
    'Group': ['A', 'B', 'C', 'D', 'E'],
    'Value': [1, 5, 4, 3, 9]
})
```

```python
plt.barh(y=df.Group, width=df.Value)
```

```python

fig, ax = plt.subplots()

# Çizgi grafiği oluştur
ax.plot(df.Value, df.Group, label='Veri Seti')

# Başlık ve etiketler ekleyin
ax.set_title('Basit Bir Çizgi Grafiği')
ax.set_xlabel('Değerler Ekseni')
ax.set_ylabel('Group Ekseni')

# Lejantı göster
ax.legend()

# Grafiği göster
plt.show()
```

```python
import matplotlib.pyplot as plt

# Veriyi oluştur
x = [1, 2, 3, 4, 5]
y = [10, 15, 7, 12, 9]

# Figür ve alt-çizim oluştur
fig, ax = plt.subplots()

# Çizgi grafiği oluştur
ax.plot(x, y, label='Veri Seti')

# Başlık ve etiketler ekleyin
ax.set_title('Basit Bir Çizgi Grafiği')
ax.set_xlabel('X Ekseni')
ax.set_ylabel('Y Ekseni')

# Lejantı göster
ax.legend()

# Grafiği göster
plt.show()
```

```python
df = df.sort_values(by=['Value'])  # 'Value' sütununa göre küçükten büyüğe doğru sıralama
my_plot = plt.barh(
    y=df.Group,
    width=df.Value,
    color=mcolors.TABLEAU_COLORS,
    edgecolor='black',
    linewidth=3
    )
plt.bar_label(container=my_plot, padding=4)  # Çubukların üzerine değer etiketlerini ekler.
#container=my_plot: Etiketlerin ekleneceği çubukları belirtir.
#padding=4: Etiketlerin çubuklardan uzaklığını belirler.
plt.show()
```

```python
tips = sns.load_dataset('tips')
#Seaborn kütüphanesinin içinde bulunan önceden tanımlanmış bir veri setini yüklemek için kullanılır. 
# tips burada bir restoran müşterilerinden alınan bir veri setini temsil eder.
```

```python
tips.head()
```

```python
tips.info()
```

```python
sns.set(style="darkgrid")
sns.barplot(
    x="total_bill",
    y="day",
    data=tips,
    color="pink"
)
plt.figure(figsize=(18, 8)) #çubuk grafiğin oluşturulduğu figürün (figure) boyutunu belirler. 'figsize' parametresi ile genişlik ve yükseklik belirlenir. 
#Bu durumda, genişlik 18 birim, yükseklik 8 birim olacak şekilde bir figür oluşturulur.
```

'tips' veri setindeki günlerin (day) her biri için cinsiyet (sex) ve total_bill sütunlarına göre ayrı çubuk grafikler oluştur. Her bir gün, alt-çizimlerde (kolonlarda) göster ve cinsiyetlere göre total_bill değerlerini karşılaştır.

```python
#Seaborn kütüphanesini kullanarak FacetGrid ve barplot özelliklerini birleştirerek bir çoklu çubuk grafiği (facetted bar plot) 
g = sns.FacetGrid(tips, col="day", height=4, aspect=.5)
g.map(sns.barplot, "sex", "total_bill", order=["Male", "Female"])
```

```python
sns.barplot(
        x="total_bill",
        y="smoker",
        hue="sex",
        data=tips,
        palette='dark:pink'
        )
plt.figure(figsize=(16, 8))
```

```python
tips.head()
```

```python
total = tips.groupby('day')['total_bill'].sum().reset_index()
# 'total' adlı yeni DataFrame, her bir gün için toplam 'total_bill' değerlerini içerir. 
#.reset_index(): Gruplama işlemi sonucu indeksler sıfırlanır ve bu toplam değerleri içeren yeni bir DataFrame oluşturulur.
```

```python
total
```

```python
bar1 = sns.barplot(x="day",  y="total_bill", data=total, color='darkblue')
```

Sigara içen müşterilerin her bir gün için toplam harcamalarını gösteren bir çubuk grafiği 

```python
smoker = tips[tips.smoker=='Yes']
```

```python
bar2 = sns.barplot(
    x='day',
    y='total_bill',
    data=smoker,
    estimator=sum, #Her bir gün için 'total_bill' değerlerinin toplamını hesaplar ve çubukların yüksekliğini bu toplamlara göre belirler.
    errorbar=None, #Hata çubuklarını kaldırır.
    color='lightblue'
)
```

Bu kod parçası, "non-smoker" ve "smoker = Yes" olmak üzere iki farklı durumu karşılaştıran yan yana çubuk grafikleri oluşturur ve bunları bir lejant ile birlikte gösterir.

```python
bar1 = sns.barplot(x="day",  y="total_bill", data=total, color='darkblue')
bar2 = sns.barplot(
    x='day',
    y='total_bill',
    data=smoker,
    estimator=sum,
    errorbar=None,
    color='lightblue'
)
top_bar = mpatches.Patch(color='darkblue', label='non-smoker') #üst çubuğu (non-smoker'ı temsil eden çubuk) belirler.
bottom_bar = mpatches.Patch(color='lightblue', label='smoker = Yes') #Lejant için kullanılacak alt çubuğu (smoker'ı temsil eden çubuk) belirler.
plt.legend(handles=[top_bar, bottom_bar]) #Lejantı oluşturur ve grafiğin sağ üst köşesine yerleştirir.
plt.show()
```

```python
df_iris = sns.load_dataset('iris')
```

```python
df_iris.head()
```

 ### histogram

```python
sns.histplot(data=df_iris, x='sepal_length', kde=True)
#Histogramın üstüne bir çekirdek yoğunluk tahmini (Kernel Density Estimate - KDE) ekler. Bu, veri setinin genel dağılımını daha yumuşak bir çizgi ile gösterir.
```

### Regresyon Gradfiği

```python
sns.regplot(
    x=df_iris['sepal_length'],
    y=df_iris['sepal_width'],
    marker= "+",
    line_kws={           #içinde çizgi rengi, saydamlık (alpha), ve kalınlık (lw) gibi özellikleri ayarlar.
        "color":"red",
        "alpha":0.6,
        "lw":5 
    },
    scatter_kws={          #renk ve marker boyutu gibi özellikleri ayarlar.
        "color":"darkblue",
        "s":50
    }
    )
```

```python
sns.lmplot(
    x='sepal_length',
    y='sepal_width',
    data=df_iris,
    hue="species"  # 'species' sütununa göre renklendir  Bu, farklı iris çiçek türlerini temsil eden renklerin kullanılmasını sağlar.
    
)

"""Bu kod parçası, 'iris' veri setindeki 'sepal_length' ve 'sepal_width' sütunları arasında 
    bir regresyon grafiği oluşturur 
    ve bu grafiği farklı iris çiçek türlerini temsil eden renklerle gösterir.
    """
```

### Kutu grafiği (box plot)

 'iris' veri setindeki çiçek türleri arasındaki 'sepal_length' değerlerinin kutu grafiğini oluşturur ve görüntüler. Renk paleti ve tema ayarları, grafiği daha estetik hale getirmek için kullanılır.

```python
sns.set_theme(style="white", palette="pastel")
sns.boxplot(x=df_iris['species'], y=df_iris['sepal_length'])
```

```python
df_iris.boxplot(column='sepal_length', by='species')
plt.show()
#by='species': Veri setini bölecek kategorik bir değişken belirler. Bu durumda 'species' yani çiçek türleri.
```
