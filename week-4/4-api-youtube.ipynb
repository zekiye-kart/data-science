## Creating workflow

```python
import requests
import pandas as pd
import time
#from config import YOUTUBE_API_KEY
```

```python
def get_video_statistics(video_id, youtube_api_key):
    # collecting video statistics
    #Bu satırlarda, belirli bir video için istatistik bilgilerini çekmek üzere YouTube API'sine bir HTTP GET isteği gönderilir ve gelen yanıt JSON formatında okunur. Bu yanıt, video ile ilgili istatistik bilgilerini içerir.
    video_stats_url = "https://www.googleapis.com/youtube/v3/videos?id="+video_id+"&part=statistics&key="+youtube_api_key
    video_stats_response = requests.get(video_stats_url).json()
    


    #İstatistik Bilgilerini Ayıklama:
    #Bu satırlarda, API'den gelen yanıttan sadece görüntüleme sayısı (viewCount) ve beğeni sayısı (likeCount) istatistikleri alınır.
    view_count = video_stats_response['items'][0]["statistics"]["viewCount"]
    like_count = video_stats_response['items'][0]["statistics"]["likeCount"]
    
    
    #İstatistik Bilgilerini Geri Döndürme:
    #Fonksiyon, alınan istatistik bilgilerini bir demet (tuple) olarak geri döndürür. Bu sayede, fonksiyonu çağırdığınız yerde bu istatistik bilgilerini kullanabilirsin
    return view_count, like_count
```

Belirli bir YouTube kanalındaki videoların bilgilerini çekerek, bu bilgileri bir Pandas DataFrame'e eklemek:

```python
def get_videos_from_channel(df, channel_id, youtube_api_key):
    # make API call
    pageToken = ""
    url = "https://www.googleapis.com/youtube/v3/search?key="+youtube_api_key+"&channelId="+channel_id+"&part=snippet,id&order=date&maxResults=1000"+pageToken
    
    #YouTube API'ye İstek Gönderme:
    response = requests.get(url).json()      # API'nin döndürdüğü yanıt JSON formatında okunur.
    
    #API Çağrısından Sonra Bekleme:
    time.sleep(3)                             #API çağrısından sonra 3 saniye bekleme süresi ekler. Bu, API'ye sık sık yapılan istekleri sınırlamak ve olası bir kısıtlamadan kaçınmak için yaygın bir uygulamadır.



    #Her Bir Video İçin İstatistikleri ve Bilgileri Çekme:
    for video in response['items']:
        if video['id']['kind'] == 'youtube#video':
            video_title = video['snippet']['title']
            video_id = video['id']['videoId']
            upload_date = video['snippet']['publishedAt']
            upload_date= str(upload_date).split("T")[0]

            view_count, like_count = get_video_statistics(video_id, youtube_api_key)  #get_video_statistics fonksiyonunu kullanarak video istatistiklerini çeker



            #Pandas DataFrame'e Bilgileri Ekleme:   her bir video için elde edilen bilgileri bir Pandas DataFrame'e ekler.
            df = df._append({
                "video_id": video_id,
                "video_title": video_title,
                "upload_date": upload_date,
                "view_count": view_count,
                "like_count": like_count
            }, 
            ignore_index = True
            )
    return df
```

```python
CHANNEL_ID = "UCLA_DiR1FfKNvjuUpBHmylQ"

df = pd.DataFrame(columns=["video_id","video_title","upload_date","view_count","like_count"]) 
youtube_data = get_videos_from_channel(df, CHANNEL_ID, YOUTUBE_API_KEY)
```
