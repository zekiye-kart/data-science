```python
import requests
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

web sitesinden veri çekme işlemi:

```python

url = "https://realpython.github.io/fake-jobs/"
page = requests.get(url)      #Belirtilen URL'den HTTP GET isteği yaparak web sayfasının içeriğini çeker.
soup = BeautifulSoup(page.content, 'html.parser')  # added .content to avoid problems with character encoding.
#requests ile çekilen sayfa içeriğini, BeautifulSoup kütüphanesi aracılığıyla işlemek için kullanılır. 'html.parser', HTML belgesini analiz etmek için BeautifulSoup'e hangi ayrıştırıcıyı kullanması gerektiğini söyler.
```

```python
type(soup)
```

```python
print(soup.prettify()) # belgenin içeriğini düzenli bir şekilde boşluklar ve girintiler ekleyerek biçimlendirir. 
```

```python
job_elements = soup.find_all("div", class_="card-content") # İlanları temsil eden div öğelerini bul
```

```python
job_elements
```

```python
for job_element in job_elements:
    title_element = job_element.find("h2", class_="title")
    company_element = job_element.find("h3", class_="company")
    location_element = job_element.find("p", class_="location")
    print(title_element)
    print(company_element)
    print(location_element)
    print("\n")
```

```python
print(f"There are total {len(job_elements)} jobs:")
print("\n")

for num, job_element in enumerate(job_elements):
    title_element = job_element.find("h2", class_="title")
    company_element = job_element.find("h3", class_="company")
    location_element = job_element.find("p", class_="location")
    print(num+1)
    print(title_element.text.strip())
    print(company_element.text.strip())
    print(location_element.text.strip())
    print("\n")
```

```python
python_jobs = soup.find_all(
    "h2", string=lambda text: "python" in text.lower()
)
#find_all(): Belirli bir HTML etiketini seçen BeautifulSoup metodudur. Bu durumda, h2 etiketini ifade eder.
#lambda text: "python" in text.lower(): Bu lambda fonksiyonu, bir metin parametresi alır (text) ve bu metinde "python" kelimesinin küçük harfle geçip geçmediğini kontrol eder. 
# text.lower() ifadesi, metindeki tüm harfleri küçük harfe dönüştürür, böylece büyük/küçük harf duyarlılığını ortadan kaldırır.

#Bu lambda fonksiyonu, True döndürdüğü durumda, yani metin içinde "python" kelimesi geçiyorsa, 
# ilgili h2 elementi seçilir ve python_jobs listesine eklenir

#Sonuç olarak, python_jobs değişkeni, HTML belgesinde bulunan ve başlığı içinde "python" kelimesini içeren tüm h2 elementlerini içerecektir. 
# Bu elementlere daha sonra erişebilir ve gerekli bilgileri çekebilirsiniz.

#soup nesnesi içinde bulunan tüm h2 başlıklarını (heading 2) seçer, 
#ancak sadece başlıkların metni içinde "python" kelimesini içerenleri filtreler.
```

```python
for job in python_jobs:
    print(job)
```

Anchor elementi < a > (anchor) HTML etiketi, web sayfalarında bağlantıları tanımlamak için kullanılır. < a> etiketi, kullanıcıları başka bir sayfaya, aynı sayfadaki belirli bir bölüme veya farklı bir web kaynağına yönlendirmek için kullanılır. Bir bağlantı oluşturulduğunda, genellikle bir metin veya resim içinde görüntülenir.

```python
for job in python_jobs:
    print(job.find_all('a'))
    #Her bir Python iş ilanının içindeki tüm anchor (<a>) elementlerini bulur ve ekrana yazdırır. 
    # Ancak, bu sadece bağlantıların kendisini, yani HTML <a> elementlerini yazdırır, 
    # içerdikleri metin veya URL'leri yazdırmaz.
```

 belirli bir filtreleme işlemi kullanarak, Python iş ilanlarını temsil eden başlık elementleri olan<   h2> elementlerinin ebeveynlerini (parent) çekmeye çalış

```python
python_job_elements = [
    h2_element.parent.parent.parent for h2_element in python_jobs
]
#<h2> elementinin üç seviye yukarıdaki bir konteynerin içinde olduğunu düşündürüyor.
```

```python
for sibling in python_job_elements[0].next_siblings: #python_job_elements[0] listesindeki ilk Python iş ilanını temsil eden öğenin sonraki kardeş öğeleri üzerinde bir döngü başlatır.
    print(sibling)  #Her bir kardeş öğeyi ekrana yazdırır. Bu, öğelerin etiketlerini ve içeriklerini içerebilir.
```

Her bir iş ilanının içinde bulunan tüm <   a> (anchor) elementlerini (links) bulur ve bu elementlerin metin içeriğini ekrana yazdırır.

```python
for job_element in python_job_elements:
    links = job_element.find_all('a') #Her bir iş ilanının içindeki tüm <a> elementlerini bulur ve bunları links adlı bir liste içine alır.
    for link in links:   # Her bir <a> elementi üzerinde bir döngü başlatır.
        print(link.text.strip()) #Her bir <a> elementinin metin içeriğini ekrana yazdırır. text.strip() metodu, metin içeriğindeki başındaki ve sonundaki boşlukları temizler.
```

```python
for job_element in python_job_elements:
    links = job_element.find_all('a')
    for link in links:
        if link.text == "Apply":
            print(link)
```

```python
python_headlines = [headline.text.strip() for headline in soup.find_all('h2') if 'Python' in headline.text]
word_counts = [len(headline.split()) for headline in python_headlines]
plt.figure(figsize=(10, 6))
plt.bar(range(len(python_headlines)), word_counts)
plt.xlabel('Başlık')
plt.ylabel('Kelime Sayısı')
plt.title('Python İçeren Başlıkların Kelime Sayısı')
plt.xticks(range(len(python_headlines)), python_headlines, rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
url = 'https://realpython.github.io/fake-jobs/'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
python_headlines = [headline.text.strip() for headline in soup.find_all('h2') if 'Python' in headline.text]
word_counts = [len(headline.split()) for headline in python_headlines]
plt.figure(figsize=(10, 6))
plt.pie(word_counts, labels=python_headlines, autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title('Python İçeren Başlıkların Kelime Sayısı')
plt.show()
```

```python
url = 'https://realpython.github.io/fake-jobs/'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
python_headlines = [headline.text.strip() for headline in soup.find_all('h2') if 'Python' in headline.text]
word_counts = [len(headline.split()) for headline in python_headlines]

plt.figure(figsize=(10, 6))
sns.violinplot(y=word_counts, color='skyblue')
plt.ylabel('Kelime Sayısı')
plt.title('Python İçeren Başlıklardaki Kelime Sayısının Violin Plot ile Dağılımı')
plt.grid(True)
plt.show()
```
