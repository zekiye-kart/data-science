```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline  # Jupyter Notebook ortamında çalışırken, matplotlib kütüphanesinin çizimlerini ve grafiklerini hücrenin içinde göstermek için kullanılır.
```

```python
x = np.arange(-5.0, 5.0, 0.1) #Bu satır, -5'ten 5'e kadar (5 dahil değil) olan değerleri 0.1 aralıklarla içeren bir NumPy dizisi oluşturur. Bu dizi, x değerlerini temsil eder.

##You can adjust the slope and intercept to verify the changes in the graph
# belirli bir eğim (slope) ve düzlem (intercept) değeri ile bir doğru oluşturur, ardından bu doğruya bir miktar rastgele gürültü ekleyerek bir veri seti oluşturur.


y = 3 + 2*(x) #Bu satır, belirli bir eğim ve düzlem kullanarak bir doğru oluşturur. Burada, eğim 2, düzlem ise 3'tür. Yapılan işlem, lineer regresyon modelinin temel formülüdür: y= eğim . x + düzlem 

y_noise = 2 * np.random.normal(size=x.size)
#np.random.normal fonksiyonunu kullanarak belirli bir boyutta (burada x dizisinin boyutu) rastgele gürültü ekler. 2 ile çarpılması, gürültü miktarını artırır.
ydata = y + y_noise
#oluşturulan doğruya rastgele gürültüyü ekleyerek bir veri seti oluşturur. ydata, gerçek değerler ile gürültü eklenmiş değerleri içeren bir dizi olur.
```

```python
plt.figure(figsize=(12,8))
plt.plot(x, ydata,  'bo')
plt.plot(x,y, 'r') 
plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable')
plt.show()
```

 belirli bir üçüncü dereceden polinom fonksiyonu ile bir doğru oluşturur ve bu doğruya bir miktar rastgele gürültü ekleyerek bir veri seti oluştur

```python
x = np.arange(-5.0, 5.0, 0.1)

##You can adjust the slope and intercept to verify the changes in the graph
y = 1*(x**3) + 1*(x**2) + 1*x + 3
y_noise = 20 * np.random.normal(size=x.size)
ydata = y + y_noise
```

```python
plt.plot(x, ydata,  'bo')
plt.plot(x,y, 'r') 
plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable')
plt.show()
```

## Quadratic
belirli bir ikinci dereceden polinom fonksiyonunu (parabol) temsil eden bir eğri oluşturur ve bu eğriye bir miktar rastgele gürültü ekleyerek bir veri seti oluştur

```python
x = np.arange(-5.0, 5.0, 0.1)

##You can adjust the slope and intercept to verify the changes in the graph

y = np.power(x,2)
y_noise = 2 * np.random.normal(size=x.size)
ydata = y + y_noise
```

```python
plt.plot(x, ydata,  'bo')
plt.plot(x,y, 'r') 
plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable')
plt.show()
```

## Exponential
Y = a+b c^x

bir NumPy dizisi olan X üzerinde eğri üzerindeki noktaları temsil eden bir NumPy dizisi Y oluşturur. Y dizisi, X dizisindeki her bir elemanın üstel fonksiyonu (e^x) tarafından hesaplanan değerleri içerir.

```python
X = np.arange(-5.0, 5.0, 0.1)

##You can adjust the slope and intercept to verify the changes in the graph

Y= np.exp(X) #X dizisindeki her bir x değeri için üstel fonksiyonu (e^x) kullanarak bir y değeri hesaplar ve bu değerleri içeren bir Y dizisi oluşturur.
```

```python
plt.plot(X,Y) 
plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable')
plt.show()
```

## Logistic / Sigmoid
Y = a + (b / 1 + c ^ (X-d))
sigmoid fonksiyonu kullanılarak bir eğri oluştur

sigmoid fonksiyonunu temsil eden bir Y dizisi oluşturur. Sigmoid fonksiyonu genellikle logistic regresyon gibi modellerde kullanılır.

```python
X = np.arange(-5.0, 5.0, 0.1)

Y = 1-4/(1+np.power(3, X-2))
```

```python
plt.plot(X,Y) 
plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable')
plt.show()
```

```python
import numpy as np
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/zekiye-kart/data-science/main/week-7/china_gdp.csv"
df = pd.read_csv(url)
```

```python
x_data, y_data = (df["Year"].values, df["Value"].values)
plt.plot(x_data, y_data, "ro")
plt.ylabel("GDP")
plt.xlabel("Year")
plt.show()
```

```python
X = np.arange(-5.0, 5.0, 0.1)
Y = 1.0 / (1.0 + np.exp(-X))   #sigmoid (logistic) fonksiyonu   exp(-X)-->euler  e^(-x)

plt.plot(X,Y) 
plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable')
plt.show()
```

```python
def sigmoid(x, Beta_1, Beta_2):
    
    """
     Özel bir sigmoid fonksiyonu tanımlar.

     Parameters:
     - x: Bağımsız değişken.
     - Beta_1, Beta_2: Sigmoid fonksiyonunun parametreleri.

     Returns:
     - y: Hesaplanan sigmoid fonksiyonunun değeri.
     """
    
    y = 1 / (1 + np.exp(-Beta_1*(x-Beta_2)))
    return y
```

```python
#belirli parametre değerleri ile sigmoid fonksiyonunu kullanarak bir tahmin yapar 
# ve bu tahmini gerçek veri noktalarıyla birlikte bir grafik üzerinde görselleştirir.

beta_1 = 0.10
beta_2 = 1990.0

#logistic function
Y_pred = sigmoid(x_data, beta_1 , beta_2)

#plot initial prediction against datapoints
plt.plot(x_data, Y_pred*15000000000000.)
plt.plot(x_data, y_data, 'ro')
#gerçek veri kırmızı, tahmin mavi
```

Aşağıdaki kod her iki veri setini de 0 ile 1 arasına ölçeklendirir, çünkü her iki veri seti de maksimum değerine bölünmektedir. Normalizasyon işlemi, özellikle makine öğrenimi modellerinde girdi verilerinin birbirine daha yakın bir aralıkta olmasını sağlamak için sıklıkla kullanılır.

```python
# Normalize our data
xdata = x_data / max(x_data)
ydata = y_data / max(y_data)
```

curve_fit fonksiyonu, veri seti üzerinde belirtilen bir model fonksiyonuna en iyi uyan parametreleri bulmaya yönelik bir optimize işlemi gerçekleştirir.

```python
from scipy.optimize import curve_fit

# curve_fit fonksiyonunu kullanarak sigmoid fonksiyonuna en iyi uyan parametreleri bulma
popt, pcov = curve_fit(sigmoid, xdata, ydata)

# Elde edilen parametreleri yazdırma
print(" beta_1 = %f, beta_2 = %f" % (popt[0], popt[1]))


#Bu kodun sonucunda ekrana yazdırılan beta_1 ve beta_2, sigmoid fonksiyonuna en iyi uyan parametrelerdir. 
# Bu parametreler, veri seti üzerindeki sigmoid eğrinin şeklini belirler.
```

```python
x = np.linspace(1960, 2015, 55)  # x dizisi belirli bir aralıktaki yılları temsil eder ve bu yılları normalize etmek için maksimum değere bölünür.
x = x/max(x)
plt.figure(figsize=(12,8))

# Eğriyi çizmek için sigmoid fonksiyonunu kullanma
y = sigmoid(x, *popt)

# Matplotlib kullanarak eğriyi ve veri noktalarını görselleştirme
plt.plot(xdata, ydata, 'ro', label='data')
plt.plot(x,y, linewidth=3.0, label='fit')
plt.legend(loc='best')
plt.ylabel('GDP')
plt.xlabel('Year')
plt.show()
```

eğitim seti üzerinde bir sigmoid modelini kurar, ardından test seti üzerinde tahmin yapar ve modelin performansını değerlendirir.

```python
# split data into train/test
msk = np.random.rand(len(df)) < 0.8   #rastgele bir maske oluşturarak veriyi %80 eğitim ve %20 test setine ayırır.
train_x = xdata[msk]
test_x = xdata[~msk]
train_y = ydata[msk]
test_y = ydata[~msk]

# Eğitim seti üzerinde sigmoid modelini oluşturma
popt, pcov = curve_fit(sigmoid, train_x, train_y)   #curve_fit fonksiyonu kullanılarak eğitim seti üzerinde sigmoid modeli oluşturulur (popt içindeki parametreler).

# Test seti üzerinde tahmin yapma
y_hat = sigmoid(test_x, *popt)

# Model performansını değerlendirme
print("Mean absolute error: %.2f" % np.mean(np.absolute(y_hat - test_y)))
print("Residual sum of squares (MSE): %.2f" % np.mean((y_hat - test_y) ** 2))
from sklearn.metrics import r2_score
print("R2-score: %.2f" % r2_score(test_y,y_hat) )
```

*popt ifadesi, bir Python özelliği olan "unpacking" (açma) işlemidir. curve_fit fonksiyonu, optimize edilmiş parametreleri içeren bir NumPy dizisi döndürür. Bu diziyi bir fonksiyona geçirirken, * operatörü ile diziyi açabilirsiniz.

popt dizisinin her elemanını sırasıyla Beta_1 ve Beta_2 parametrelerine yerleştirir ve sigmoid fonksiyonunu çağırır. Bu şekilde, optimize edilmiş parametreleri doğrudan fonksiyona geçirebilirsiniz.
